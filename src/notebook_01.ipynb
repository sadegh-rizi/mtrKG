```python
import networkx as nx
import matplotlib.pyplot as plt

# 1. Initialize a Directed Graph
G = nx.DiGraph()

# 2. Define the Nodes and their specific colors (representing biological layers)
nodes = [
    ("Esketamine", {"color": "#99ccff"}),            # Pharmacology (Blue)
    ("Amantadine", {"color": "#99ccff"}),            # Pharmacology (Blue)
    ("GRIN3B\n(NMDA Receptor)", {"color": "#99cc99"}), # Gene/Protein (Green)
    ("Sphingolipid\nRatio", {"color": "#ffcc99"}),     # Metabolomics (Orange)
    ("rs4147929\n(Variant)", {"color": "#d9d9d9"}),    # Genomics (Grey)
    ("Alzheimer's\nDisease", {"color": "#ff9999"})     # Clinical Phenotype (Red)
]
G.add_nodes_from(nodes)

# 3. Define the Edges (The biological causality links)
G.add_edge("rs4147929\n(Variant)", "Alzheimer's\nDisease", label="GWAS Link")
G.add_edge("rs4147929\n(Variant)", "Sphingolipid\nRatio", label="rQTL Link")
G.add_edge("GRIN3B\n(NMDA Receptor)", "Sphingolipid\nRatio", label="MR Causal Driver\n(p = 2.3e-19)")
G.add_edge("Esketamine", "GRIN3B\n(NMDA Receptor)", label="Inhibits")
G.add_edge("Amantadine", "GRIN3B\n(NMDA Receptor)", label="Inhibits")

# 4. Extract assigned colors for plotting
colors = [nx.get_node_attributes(G, 'color')[node] for node in G.nodes()]

# 5. Manually position the nodes for a beautiful Left-to-Right narrative flow
pos = {
    "Esketamine": (0, 0.8),
    "Amantadine": (0, -0.8),
    "GRIN3B\n(NMDA Receptor)": (1.5, 0),
    "Sphingolipid\nRatio": (3.5, 0),
    "rs4147929\n(Variant)": (5.5, 0),
    "Alzheimer's\nDisease": (7.5, 0)
}

# 6. Initialize the Matplotlib Canvas
fig, ax = plt.subplots(figsize=(14, 5))

# Draw the Nodes
nx.draw_networkx_nodes(
    G, pos,
    node_color=colors,
    node_size=4000,
    edgecolors="black",
    linewidths=1.5,
    ax=ax
)

# Draw the Node text
nx.draw_networkx_labels(
    G, pos,
    font_size=10,
    font_weight="bold",
    ax=ax
)

# Draw the Edges (Arrows pointing in the direction of biological effect)
nx.draw_networkx_edges(
    G, pos,
    edge_color="gray",
    arrows=True,
    arrowsize=20,
    node_size=4000,
    width=2,
    ax=ax
)

# Draw the Edge text
edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels=edge_labels,
    font_color="darkred",
    font_size=9,
    font_weight="bold",
    label_pos=0.5,
    ax=ax
)

# 7. Clean up and Save
ax.axis("off")
plt.margins(0.1)

# Save as a high-resolution PNG perfect for a PowerPoint/LaTeX PDF!
plt.savefig("alzheimers_pathway.png", dpi=300, bbox_inches='tight')
```

```python
import qrcode
import os

def links_to_qrcodes(links, output_directory="qr_codes"):
    """
    Converts a list of links into QR code images and saves them to a directory.

    Args:
        links (list): A list of strings containing the URLs.
        output_directory (str): The folder where images will be saved.
    """
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    for index, link in enumerate(links):
        # Configure the QR code properties
        qr = qrcode.QRCode(
            version=1, # Controls the size of the QR Code (1 to 40)
            error_correction=qrcode.constants.ERROR_CORRECT_L, # About 7% or less errors can be corrected
            box_size=10, # Size of each 'box' or 'pixel' in the QR code
            border=4, # Thickness of the border (minimum is 4)
        )

        # Add the link data to the QR code
        qr.add_data(link)
        qr.make(fit=True)

        # Generate the image
        img = qr.make_image(fill_color="black", back_color="white")

        # Create a safe filename and save the image
        filename = f"qr_link_{index + 1}.png"
        filepath = os.path.join(output_directory, filename)

        img.save(filepath)
        print(f"Saved: {filepath} -> {link}")

# --- Example Usage ---
my_links = [
    "https://github.com/sadegh-rizi/mtrKG",

]

links_to_qrcodes(my_links)
```
