```python
# src/validate_graph.py
from pyshacl import validate
from rdflib import Graph

def run_quality_check(data_graph_path, shacl_shape_path):
    print(f"Running Data Quality Assessment on {data_graph_path}...")

    # Load the data and the shapes
    data_g = Graph().parse(data_graph_path, format="turtle")
    shacl_g = Graph().parse(shacl_shape_path, format="turtle")

    # Run PySHACL
    conforms, report_graph, report_text = validate(
        data_g,
        shacl_graph=shacl_g,
        inference='rdfs', # Uses the RDFS reasoner to respect your subClassOf rules!
        abort_on_first=False,
        meta_shacl=False,
        debug=False
    )

    if conforms:
        print("✅ SUCCESS: Your Knowledge Graph passed all quality constraints!")
    else:
        print("❌ QUALITY ISSUES DETECTED:")
        print(report_text)

if __name__ == "__main__":
    # Ensure the paths match your folder structure
    run_quality_check("../output/mtrKG.ttl", "../validation/schema_validation.ttl")
```
